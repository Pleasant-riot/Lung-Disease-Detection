og - mask, zip --> bitwise -> new folder


```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
def rv_3ch(og_img_path, mask_path, size):
    
    #module import
    import cv2
    
    # im_read
    og_img = cv2.imread(og_img_path)
    mask = cv2.imread(mask_path)
    
    # clahe
    clahe = cv2.createCLAHE(5.0,(12,12))
    
    ## og_img_convert
    cvt_img = cv2.resize(og_img, size)
    cvt_img = cv2.cvtColor(cvt_img,cv2.COLOR_BGR2GRAY)
    cvt_img = clahe.apply(cvt_img)  # 1 channel
    cvt_img = np.repeat(cvt_img[:,:,np.newaxis],3,-1)

    ## mask
    kernel = np.ones((5,5))
    cvt_mask = cv2.dilate(mask,kernel)
    cvt_mask = cv2.resize(mask, (size))
    cvt_mask = cv2.cvtColor(cvt_mask, cv2.COLOR_BGR2GRAY)

    ## remove
    remove = cv2.bitwise_and(cvt_img, cvt_img, mask = cvt_mask)
    
    return remove
```

```python
from tqdm import tqdm
comp_task = []
error = []

save_path = '/content/drive/MyDrive/3차플젝drive/KAGGLE_UNET/BITWISE/'

for i,j in tqdm(zip(og_img_list,unet_mask_list)):
    try:
        fname = img_retext(i)
        format = '.jpg'

        remove = rv_3ch(i,j,(1024,1024))
    
        plt.imsave(save_path+fname+format,remove)
        comp_task.append(i)

    except Exception as e:
        print(e)
        error.append(i)
```
