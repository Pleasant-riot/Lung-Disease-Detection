```python

import cv2
import matplotlib.pyplot as plt


path= '/content/drive/MyDrive/3차플젝drive/sample_img.png'
img= cv2.imread(path)
# gray scale
img_gray  = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# threshold
_, img_binary  = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY)
# contour 잡기 (윤곽선은 투명)
contours, hierarchy = cv2.findContours(img_binary , cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image=img, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)

# see the results
plt.imshow(img)
```
